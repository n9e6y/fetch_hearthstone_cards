# handling json

```python
# to different csv based type 
unique_types = flat_data['type'].unique()

output_directory = 'output'
os.makedirs(output_directory, exist_ok=True)

for unique_type in unique_types:
    filtered_data = flat_data[flat_data['type'] == unique_type]
    
    if unique_type == '':
        filename = f'{output_directory}/type_empty.csv'
    else:
        filename = f'{output_directory}/type_{unique_type.replace(" ", "_")}.csv'
    
    filtered_data.to_csv(filename, index=False)
    
    print(f'Saved {filename}')
```
